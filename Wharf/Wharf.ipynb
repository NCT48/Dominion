```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wharf_data = pd.read_csv('./TestProgramWharf.csv', header=0, index_col=0)
wharf_data.head()
```

```python
wharf_data34 = wharf_data[wharf_data["Hund"].str.startswith("屋屋銅銅銅/", "屋銅銅銅銅/")]
wharf_data25 = wharf_data[wharf_data["Hund"].str.startswith("銅銅銅銅銅/", "屋屋屋銅銅/")]
```

```python
wharf_data34_des = wharf_data34["Turn"].describe()
wharf_data25_des = wharf_data25["Turn"].describe()

wharf_data25_des, wharf_data34_des
```

```python
plt.hist([wharf_data25["Turn"],wharf_data34["Turn"]],bins=9, range=(9.5,18.5), density=True, rwidth=0.5)
```

```python
wharf_data_sizumi = wharf_data[wharf_data["Hund"].str.find("手") > 24]
wharf_data_sizumazu = wharf_data[wharf_data["Hund"].str.find("手") < 23]
```

```python
wharf_data_sizumi_des = wharf_data_sizumi["Turn"].describe()
wharf_data_sizumazu_des = wharf_data_sizumazu["Turn"].describe()

wharf_data_sizumi_des, wharf_data_sizumazu_des
```

```python
plt.hist([wharf_data_sizumi["Turn"],wharf_data_sizumazu["Turn"]],bins=9, range=(9.5,18.5), density=True, rwidth=0.5)
```

```python
wharf_data_sizumi25 = wharf_data25[wharf_data25["Hund"].str.find("手") > 24]
wharf_data_sizumazu25 = wharf_data25[wharf_data25["Hund"].str.find("手") < 23]
```

```python
plt.hist([wharf_data_sizumi25["Turn"],wharf_data_sizumazu25["Turn"]],bins=9, range=(9.5,18.5), density=True, rwidth=0.5)
```
